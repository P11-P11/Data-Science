```python
def get_goles_eventos(players_df):
  import re
  eventos = players_df.dropna(subset=['Event'])

  patron = r'^[GPW]\d+\'$'

  eventos = eventos[eventos['Event'].str.match(patron, na=False)]

  eventos = eventos['Event'].values

  return np.array([re.search(r'\d+', evento).group() for evento in eventos])
```

```python
eventos = get_goles_eventos(players_df)
```

```python
def goles_por_minuto(eventos):
  res = np.zeros(125)
  for i in range(len(eventos)):
    res[int(eventos[i])] += 1
  return res
```

```python
eventos = goles_por_minuto(eventos)
```

```python
plt.hist(range(len(eventos)), bins=np.arange(len(eventos)+1), weights=eventos, edgecolor='black')

plt.title('Distribución de Goles a lo largo de un Partido')
plt.xlabel('Minutos')
plt.ylabel('Cantidad de Goles')
plt.xticks(np.arange(0, len(eventos), 10))
plt.show()
```

```python
def goles_por_minuto(players_df):

  eventos = players_df[['MatchID','Event']]

  eventos = eventos.dropna(subset=['Event'])

  eventos = eventos.groupby('MatchID')['Event'].apply(list).reset_index()

  evento_lista = eventos['Event'].values

  import re

  patron = r'^[GPW](\d+)\''

  res = []

  for evento in evento_lista:
    lista_aux = pd.Series(evento).explode().str.split(' ').explode().to_numpy()
    #print(lista_aux)
    res_aux = []
    for i in lista_aux:
      aux = re.findall(patron, i)

      if aux !=[] and int(aux[0]) not in res_aux:
        res_aux.append(int(aux[0]))
    if res_aux != []:
      res.append(sorted(res_aux))

  return res




```

```python
def t_primerGol(players_df):
  goles = goles_por_minuto(players_df)
  res = []
  for gol in goles:
    res.append(gol[0])
  return np.array(res)
```

```python
tiempos = t_primerGol(players_df)
```

```python

sns.histplot(data=tiempos, bins=range(0, max(tiempos) + 5, 5), kde=True, edgecolor='black')

plt.title('Distribución de Tiempo hasta el Primer Gol')
plt.xlabel('Tiempo hasta el Primer Gol (minutos)')
plt.ylabel('Cantidad de Partidos')

plt.xticks(rotation=45)

plt.show()
```

```python
def delta_tiempos(players_df):
  goles = goles_por_minuto(players_df)
  res = []
  for gol in goles:
    if len(gol) > 1:
      for i in range(len(gol) - 1):
        res.append(gol[i+1] - gol[i])
  return np.array(res)
```

```python
delta_tiempo = delta_tiempos(players_df)
```

```python

sns.histplot(data=delta_tiempo, bins=range(0, max(delta_tiempo) + 5, 5), kde=True, edgecolor='black')

plt.title('Distribución de Tiempo hasta el Primer Gol')
plt.xlabel('Delta tiempo entre goles')
plt.ylabel('Cantidad de Partidos')

plt.xticks(rotation=45)

plt.show()
```
